````r
flights_sml <- select(flights, year:day, ends_with("delay"),
                     air_time, distance)
````

````r
select(flights, time_hour, air_time, everything())
````

![images](images/slides15.png)

````r
mutate(flights_sml,
      gain=arr_delay - dep_delay,
      speed=distance / air_time * 60,
      gain_per_hour = gain / (air_time / 60))
````

````r
transmute(flights_sml,
      gain=arr_delay - dep_delay,
      speed=distance / air_time * 60,
      gain_per_hour = gain / (air_time / 60))
````

````r
by_day <- group_by(flights, year, month, day)
summarize(by_day, mean_distance = mean(distance, na.rm = TRUE),
         mean_arr_delay = mean(arr_delay, na.rm = TRUE))
````

## A Data Anaysis Example

Task: relationship between distance and average delay for each location.

````r
by_dest <- group_by(flights, dest)
delay <- summarize(by_dest, 
                  count = n(),
                  dist = mean(distance, na.rm = TRUE),
                  delay = mean(arr_delay, na.rm = TRUE))
delay <- filter(delay, count > 20, dest != "HNL")
````

````r
ggplot(delay,
      aes(x=dist, y=delay)) +
      geom_point(aes(size=count), 
                alpha=1/3) +
      geom_smooth(se = FALSE)
````

It looks like delays increase with distance up to ~750 miles and then decrease.

![images](images/slides19.png)

````r
delays <- flights %>%
  filter(!is.na(dep_delay), !is.na(arr_delay)) %>%
  group_by(tailnum) %>%
  summarize(delay = mean(arr_delay) )

ggplot(data=delays, mapping=aes(x=delay)) + geom_freqpoly(binwidth=10)
````
